```python
# Construa um programa que calcule: Quanto devemos cobrar em um projeto de  programação se trabalhamos 8h por dia, demoramos 15 dias para fazer o projeto e cobramos R$100/h?
# Qual o resultado que eu quero? Quanto eu vou cobrar pelo projeto
# O que eu preciso de info p/ chegar ao resultado? Quanto tempo vou trabalhar no projeto (horas), quanto custa minha hora de trabalho
# Qual o passo a passo?
#Passo 1: Descobrir quantas hrs vou trabalhar.
#Passo 2: Descobrir quanto custa 1h de trabalho e multiplica pelo passo 1.
#Passo 3: Exibir o custo do projeto.

#Passo1:
horas_por_dia = 8
dias_trabalhados = 15
custo_hora = 100

#Passo 2:
horas_trabalhadas = horas_por_dia * dias_trabalhados
custo_projeto = horas_trabalhadas * custo_hora

#Passo 3
print(custo_projeto)
```

```python
#Construa um programa que calcule:
#Quanto custa encher o tanque de um carro que tem 50L de capacidade, está com 20L de combustível
#atualmente e o custo do combustível é de R$5,80/L

# Qual o resultado que eu quero? Qual o custo para encher o tanque
# O que eu preciso de info p/ chegar ao resultado?
# Quanto falta para encher o tanque, quanto já tem de combustível, o preço/L

#Passo 1: Descobrir quanto já tem no tanque e quanto falta
#Passo 2: Descobrir quanto falta e multiplicar pelo preço do L
#Passo 3: Exibir o custo para encher o tanque

#Passo 1
capacidade_total = 50
combustivel_atual = 20
preco_combustivel = 5.8

combustivel_falta = capacidade_total - combustivel_atual
custo_encher_tanque = combustivel_falta * preco_combustivel

#Passo 3
print(custo_encher_tanque)
```

```python
#Construa um programa que calcule: Qual deve ser o bônus do funcionário?
#Se a empresa bateu a meta de 10000 unidades,
#E se ele vendeu mais de 1000 unidades, o bonus tem que ser R$250
#Caso contrário, o bônus tem que ser R$ 50.
#Se a empresa não bateu a meta de 10000 vendas, o bônus dele tem que ser 0.

# Qual o resultado que eu quero? Qual vai ser o bônus do funcionário

# O que eu preciso de info p/ chegar ao resultado?
# Quanto ele e a empresa venderam
# Qual a meta da empresa e do funcionário
# Quanto ele ganha de bonus se bateu, quanto ganha se não bateu

#Qual o passo a passo?

#Passo 1: Definir quanto o funcionário vendeu e qual a meta de vendas
#Passo 2: Definir quanto a empresa vendeu e qual a meta de vendas
#Passo 3: Analisar se a empresa bateu a meta
#Passo 4: Se bateu a meta, analisa se ele vendeu mais do que a meta
#Passo 5: Define o bonus quando bate a meta (da empresa e do funcionário)
#Passo 6: Caso contrário do passo 4, definir bônus quando não bate a meta do funcionário
#Passo 7: Caso contrário do passo 3, define o bônus como 0
#Passo 8: Exibir bônus

#Passo 1 e 2
venda_fun = 1100
meta_fun = 1000
venda_emp = 11000
meta_emp = 10000

if venda_emp > meta_emp:
  if venda_fun > meta_fun:
    bonus = 250
  else:
    bonus = 50
else:
  bonus = 0
print(bonus)
```

```python
#Construa um programa que calcule: Qual seria o salário de um funcionário após
#10 anos se todo anos ele ganhasse 10% de aumento, salário inicial de R$2000

#Qual o resultado que você quer? Saber o salário final do funcionário
#Quais informações você precisa? Qual o aumento que ele recebe, quantos anos
#ele vai receber aumento e qual o salário inicial

#Qual o passo a passo?
#Passo 1: Definir as informações que temos
#Passo 2: Calcular o aumento salarial em 1 ano
#Passo 3: Repetir esse calculo para 10 anos
#Passo 4: Exibir o resultado final do salário


salario = 2000
aumento = 0.1
tempo = 10

for i in range (tempo):
    salario = salario * (1 + aumento)
print(salario)

```

```python
#Construa um programa que calcule: quanto tempo demora para esse funcionário chegar em um salário de R$10000?
# até ser satisfeita vai usar o WHILE

#Qual o resultado que você quer? Quanto tempo demora pra receber 10000
#Quais informações você precisa? Qual o aumento que ele recebe
# Qual o salário inicial
#Qual a meta que ele quer bater
#Algo para rastrear quanto tempo passou


#Qual o passo a passo?
#Passo 1: Definir as informações que temos
#Passo 2: Calcular 1 ano de aumento
#Passo 3: Adaptar o programa para rodar enquanto o salário for menor que a meta
#Passo 4: Exibir o resultado final de quantos anos foram necessários

#Passo 1
salario = 2000
aumento = 0.1
tempo = 0
meta = 10000

while salario < meta:
  salario = salario * (1 + aumento)
  tempo = tempo + 1
print (tempo)

```
